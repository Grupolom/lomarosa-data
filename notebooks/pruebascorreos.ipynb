```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from datetime import datetime

# ============================================
# CONFIGURACI√ìN - CAMBIA ESTOS VALORES
# ============================================
EMAIL_REMITENTE = 'datascientist@grupolom.com'  # Tu correo de Grupo LOM
PASSWORD = 'tu_contrase√±a_aqui'  # Tu contrase√±a de Outlook

# Correo de prueba (puede ser el tuyo mismo)
EMAIL_PRUEBA = 'datascientist@grupolom.com'  # Cambia esto por el correo donde quieres recibir la prueba

# Configuraci√≥n SMTP de Office 365
SMTP_SERVER = 'smtp.office365.com'
SMTP_PORT = 587

# ============================================
# SCRIPT DE PRUEBA
# ============================================
print("="*70)
print("üß™ PRUEBA DE ENV√çO DE CORREO - OUTLOOK OFFICE 365")
print("="*70)
print(f"\nüìß Correo remitente: {EMAIL_REMITENTE}")
print(f"üìß Correo de prueba: {EMAIL_PRUEBA}")
print(f"üåê Servidor SMTP: {SMTP_SERVER}:{SMTP_PORT}")
print("\n" + "="*70)

try:
    # PASO 1: Conectar al servidor SMTP
    print("\nüîÑ Paso 1/5: Conectando al servidor SMTP de Office 365...")
    server = smtplib.SMTP(SMTP_SERVER, SMTP_PORT, timeout=10)
    server.set_debuglevel(0)  # Cambiar a 1 para ver m√°s detalles
    print("   ‚úÖ Conexi√≥n establecida")
    
    # PASO 2: Identificarse con el servidor
    print("\nüîÑ Paso 2/5: Identific√°ndose con el servidor (EHLO)...")
    server.ehlo()
    print("   ‚úÖ Identificaci√≥n exitosa")
    
    # PASO 3: Iniciar encriptaci√≥n TLS
    print("\nüîÑ Paso 3/5: Iniciando encriptaci√≥n TLS...")
    server.starttls()
    server.ehlo()  # Identificarse de nuevo despu√©s de TLS
    print("   ‚úÖ Conexi√≥n segura establecida")
    
    # PASO 4: Autenticar con usuario y contrase√±a
    print("\nüîÑ Paso 4/5: Autenticando con credenciales...")
    server.login(EMAIL_REMITENTE, PASSWORD)
    print("   ‚úÖ Autenticaci√≥n exitosa")
    
    # PASO 5: Crear y enviar el correo de prueba
    print("\nüîÑ Paso 5/5: Creando y enviando correo de prueba...")
    
    # Crear el mensaje
    mensaje = MIMEMultipart('alternative')
    mensaje['From'] = EMAIL_REMITENTE
    mensaje['To'] = EMAIL_PRUEBA
    mensaje['Subject'] = 'üß™ Prueba de Env√≠o SMTP - Outlook Office 365'
    
    # Cuerpo del mensaje en texto plano
    texto_plano = f"""
¬°Prueba Exitosa!

Este es un correo de prueba enviado desde Python usando SMTP de Office 365.

Detalles de la prueba:
- Fecha y hora: {datetime.now().strftime('%d/%m/%Y %H:%M:%S')}
- Servidor SMTP: {SMTP_SERVER}
- Puerto: {SMTP_PORT}
- Correo remitente: {EMAIL_REMITENTE}

Si recibes este correo, significa que tu configuraci√≥n SMTP est√° funcionando correctamente 
y puedes proceder con el sistema de recordatorios de facturas.

‚úÖ TODO LISTO PARA AUTOMATIZAR!

---
Enviado desde Python con smtplib
    """
    
    # Cuerpo del mensaje en HTML (m√°s bonito)
    html = f"""
    <html>
        <head></head>
        <body style="font-family: Arial, sans-serif; color: #333; padding: 20px;">
            <div style="max-width: 600px; margin: 0 auto; border: 2px solid #28a745; border-radius: 10px; padding: 30px; background-color: #f8f9fa;">
                <h1 style="color: #28a745; text-align: center;">‚úÖ ¬°Prueba Exitosa!</h1>
                
                <p style="font-size: 16px; line-height: 1.6;">
                    Este es un correo de prueba enviado desde <strong>Python</strong> usando el 
                    servidor SMTP de <strong>Office 365</strong>.
                </p>
                
                <div style="background-color: white; padding: 20px; border-radius: 5px; margin: 20px 0;">
                    <h3 style="color: #0066cc; margin-top: 0;">üìä Detalles de la Prueba</h3>
                    <table style="width: 100%; border-collapse: collapse;">
                        <tr>
                            <td style="padding: 8px; border-bottom: 1px solid #ddd;"><strong>Fecha y hora:</strong></td>
                            <td style="padding: 8px; border-bottom: 1px solid #ddd;">{datetime.now().strftime('%d/%m/%Y %H:%M:%S')}</td>
                        </tr>
                        <tr>
                            <td style="padding: 8px; border-bottom: 1px solid #ddd;"><strong>Servidor SMTP:</strong></td>
                            <td style="padding: 8px; border-bottom: 1px solid #ddd;">{SMTP_SERVER}</td>
                        </tr>
                        <tr>
                            <td style="padding: 8px; border-bottom: 1px solid #ddd;"><strong>Puerto:</strong></td>
                            <td style="padding: 8px; border-bottom: 1px solid #ddd;">{SMTP_PORT} (TLS)</td>
                        </tr>
                        <tr>
                            <td style="padding: 8px;"><strong>Correo remitente:</strong></td>
                            <td style="padding: 8px;">{EMAIL_REMITENTE}</td>
                        </tr>
                    </table>
                </div>
                
                <div style="background-color: #d4edda; border-left: 4px solid #28a745; padding: 15px; border-radius: 5px;">
                    <p style="margin: 0; font-size: 14px;">
                        <strong>‚úÖ Si recibes este correo</strong>, significa que tu configuraci√≥n SMTP 
                        est√° funcionando correctamente y puedes proceder con el sistema de recordatorios 
                        de facturas.
                    </p>
                </div>
                
                <h2 style="text-align: center; color: #28a745; margin-top: 30px;">
                    üöÄ TODO LISTO PARA AUTOMATIZAR!
                </h2>
                
                <p style="text-align: center; color: #666; font-size: 12px; margin-top: 30px;">
                    Enviado desde Python con smtplib
                </p>
            </div>
        </body>
    </html>
    """
    
    # Adjuntar ambas versiones
    parte_texto = MIMEText(texto_plano, 'plain', 'utf-8')
    parte_html = MIMEText(html, 'html', 'utf-8')
    mensaje.attach(parte_texto)
    mensaje.attach(parte_html)
    
    # Enviar el correo
    server.sendmail(EMAIL_REMITENTE, EMAIL_PRUEBA, mensaje.as_string())
    print("   ‚úÖ Correo enviado exitosamente")
    
    # Cerrar la conexi√≥n
    server.quit()
    print("\nüîÑ Cerrando conexi√≥n...")
    print("   ‚úÖ Conexi√≥n cerrada correctamente")
    
    # RESULTADO FINAL
    print("\n" + "="*70)
    print("üéâ ¬°√âXITO TOTAL! üéâ")
    print("="*70)
    print("\n‚úÖ Tu cuenta de Outlook puede enviar correos perfectamente")
    print(f"üì¨ Revisa tu bandeja de entrada en: {EMAIL_PRUEBA}")
    print("\nüí° Pr√≥ximos pasos:")
    print("   1. Verifica que recibiste el correo")
    print("   2. Si lleg√≥ correctamente, puedes proceder con el sistema completo")
    print("   3. Si no lleg√≥, revisa la carpeta de spam")
    print("\n" + "="*70)

except smtplib.SMTPAuthenticationError as e:
    print("\n" + "="*70)
    print("‚ùå ERROR DE AUTENTICACI√ìN")
    print("="*70)
    print(f"\nC√≥digo de error: {e.smtp_code}")
    print(f"Mensaje: {e.smtp_error.decode()}")
    
    print("\nüí° POSIBLES SOLUCIONES:")
    print("\n1Ô∏è‚É£  CONTRASE√ëA INCORRECTA")
    print("   ‚Ä¢ Verifica que escribiste bien tu contrase√±a")
    print("   ‚Ä¢ Aseg√∫rate de no tener espacios al inicio o final")
    
    print("\n2Ô∏è‚É£  SMTP AUTH DESHABILITADO")
    print("   ‚Ä¢ Tu cuenta puede tener SMTP AUTH deshabilitado")
    print("   ‚Ä¢ Necesitas pedirle al admin de IT que ejecute este comando:")
    print(f"   Set-CASMailbox -Identity {EMAIL_REMITENTE} -SmtpClientAuthenticationDisabled $false")
    
    print("\n3Ô∏è‚É£  AUTENTICACI√ìN MULTIFACTOR (MFA)")
    print("   ‚Ä¢ Si tu cuenta tiene MFA, necesitas una contrase√±a de aplicaci√≥n")
    print("   ‚Ä¢ Ve a: https://mysignins.microsoft.com/security-info")
    print("   ‚Ä¢ Crea una contrase√±a de aplicaci√≥n y √∫sala en lugar de tu contrase√±a normal")
    
    print("\n" + "="*70)

except smtplib.SMTPConnectError as e:
    print("\n‚ùå ERROR DE CONEXI√ìN")
    print(f"No se pudo conectar al servidor {SMTP_SERVER}:{SMTP_PORT}")
    print("Verifica tu conexi√≥n a internet")

except smtplib.SMTPException as e:
    print("\n‚ùå ERROR SMTP GENERAL")
    print(f"Error: {e}")

except Exception as e:
    print("\n‚ùå ERROR INESPERADO")
    print(f"Tipo de error: {type(e).__name__}")
    print(f"Mensaje: {e}")
    
    print("\nüí° Informaci√≥n adicional para depuraci√≥n:")
    import traceback
    traceback.print_exc()

finally:
    print("\nüèÅ Prueba finalizada")
```
